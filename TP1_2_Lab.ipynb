```python
# calculos matemáticos de una variable (min, max, media)

mi=np.min(t2['birth_mortal'],axis=0) #calcula el valor mínimo a lo largo de las filas (axis=0).
# Es decir, busca el mínimo de la columna "birth_mortal"

print("El mínimo es: ",round(mi,2))

md=np.median(t2['birth_mortal'],axis=0)
print("La mediana es:",round(md,2))

mx=np.max(t2['birth_mortal'],axis=0)
print("El máximo es:", round(mx,2))
```

```python
# Usar f-string (formatted string literals) formatear cadenas en Python

mn=np.mean(t2['birth_mortal'],axis=0) # calcula el promedio
rn=round(mn,2) # redondea a 2 digitos.

print(f"El promedio es: {mn}, cuya cifra a 2 digitos es={rn}")
```

```python
# Calcular indicadores estadísticos de la variable "poverty": min, Q1,Q2,Q3, IQR, max, mediana, desvio estandard
#+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

min=t2['poverty'].min()
Q1=t2['poverty'].quantile(0.25) # 25%
Q2=t2['poverty'].quantile(0.50) # 50%
Q3=t2['poverty'].quantile(0.75) # 75%
IQR=Q3-Q1 # Threshold #
max=t2['poverty'].max()
mu=t2['poverty'].mean()
sd=t2['poverty'].std()

Value=[min,Q1,Q2,Q3,IQR,max, mu,sd]
Col=["min","25%","50%","75%","IQR","max","mean", "st_dev"]
Estadistics=pd.DataFrame(Value,Col)
Estadistics
```

```python
# Gráfico de distribución para cada variable numérica
#+++++++++++++++++++++++++++++++++++++++++++++++++++++

# Ajustar número de subplots en función del número de columnas
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(9, 9))
axes = axes.flat
columnas_numeric = t2.select_dtypes(include=['float64', 'int']).columns

for i, colum in enumerate(columnas_numeric):
    sns.histplot(
        data     = t2,
        x        = colum,
        stat     = "count",
        kde      = True,
        color    = (list(plt.rcParams['axes.prop_cycle'])*2)[i]["color"],
        line_kws = {'linewidth': 2},
        alpha    = 0.3,
        ax       = axes[i]
    )
    axes[i].set_title(colum, fontsize = 7, fontweight = "bold")
    axes[i].tick_params(labelsize = 6)
    axes[i].set_xlabel("")
    
    
fig.tight_layout()
plt.subplots_adjust(top = 0.9)
fig.suptitle('Distribución de variables numéricas', fontsize = 10, fontweight = "bold");
```

```python
# Crear una variable categórica cuyas categorias son: "low", "middle","high".
#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

#  Las condiciones son:
# q1= cuantil 25%
# q2= cuantil 50%

# Calculo de cuantiles
q1=t2['poverty'].quantile(0.25)
q2=t2['poverty'].quantile(0.50)

q1=7.472753782258431
q2=9.142439505726308

# 'poverty_index' <q1, asignar la catergoria "Low"
# q1<='poverty_index'<q2, asignar la catergoria "middle"
# 'poverty_index' >=q2, asignar la catergoria "High"

# Crea columna vacia
t2['poverty_index']=''

#establece condicionales para col. vacia
t2.loc[t2['poverty']<q1,'poverty_index'] ='Low'
t2.loc[(t2['poverty']>=q1)&(t2['poverty']<q2),'poverty_index'] ='Middle'
t2.loc[t2['poverty']>=q2,'poverty_index'] ='High'
t2
```

```python
# tabla de frecuencia de una variable categórica
#+++++++++++++++++++++++++++++++++++++++++++++++

x=t2['poverty_index'].value_counts()
x
```

```python
# graficar "poverty_index"

# Generar un array : vector
x1=np.array(["High","Middle","Low"])
y1=np.array([11,6,5])
print(x1)
print(y1)

#Mostrar una tabla (Data Frame) con x,y
k=pd.DataFrame({"categoria":x1, "frec.":y1})
print(k)

# graficar una curva
plt.bar(x1,y1,color=["red","green","blue"])
plt.title("Poverty_rate")
plt.ylabel(" quantity_province")
plt.xlabel("Level_poverty")

plt.show()
```

```python
# Gráfico de distribución de cada variable cualitativa
#++++++++++++++++++++++++++++++++++++++++++++++++++++++

# Ajustar número de subplots en función del número de columnas
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(9, 6))
axes = axes.flat
columnas_object = t2.select_dtypes(include=['object']).columns

for i, colum in enumerate(columnas_object):
    t2[colum].value_counts().plot.barh(ax = axes[i])
    axes[i].set_title(colum, fontsize = 7, fontweight = "bold")
    axes[i].tick_params(labelsize = 6)
    axes[i].set_xlabel("")

# Se eliminan los axes vacíos
for i in []:
    fig.delaxes(axes[i])
    
fig.tight_layout()
plt.subplots_adjust(top=0.9)
fig.suptitle('Distribución variables cualitativas',
             fontsize = 10, fontweight = "bold");
```
